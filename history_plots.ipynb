```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

import seaborn as sns

import numpy as np

matplotlib.rc('xtick', labelsize=10)     
matplotlib.rc('ytick', labelsize=10)
matplotlib.rc('axes', titlesize=10)
matplotlib.rc('axes', labelsize=10)
plt.rcParams["font.family"] = "serif"

sns.set_style('darkgrid')
```

```python
experiments = ['CERN I','CERN II', 'CERN III', 'BNL E821']
results = np.array([1162e-6, 11661e-7, 1165924e-9, 116592089e-11])
errors =np.array([5e-6, 3.1e-7, 8.5e-9, 63e-11])

fig1, ax1 = plt.subplots(1,1)

ax1.errorbar(range(3), results[1:], yerr=errors[1:], fmt='.', capsize=4, markersize= 10)
ax1.set_xticks(range(3))
ax1.set_xticklabels(experiments[1:])

fig1.set_size_inches(6,3)
fig1.tight_layout()
```
