```python
import numpy as np
import matplotlib.pyplot as plt
import trfp
```

```python
station_phis = np.array(trfp.STATION_BARCODE_PHI)/180*np.pi
stations = [str(st) for st in range(72)]
print stations
```

```python
fig, ax = plt.subplots(1,1)

ax = plt.subplot(1,1,1,projection='polar')


plt.polar(np.linspace(0,2*np.pi, 1000), np.full(1000,661.2), color='black')
plt.polar(np.linspace(0,2*np.pi, 1000), np.full(1000,761.2), color='black')

for st in range(72):
    plt.text(station_phis[st], 711.2, stations[st],
             horizontalalignment='center',
             verticalalignment='center',
             fontdict={'fontsize':8, 'family':'serif'})

ax.set_yticks([])
ax.set_theta_zero_location("N")
ax.set_theta_direction(-1)

ax.set_title('Fixed Probe Station Positions \n', fontdict={'fontsize':12, 'family':'serif'})

fig.set_size_inches(6,6)
fig.tight_layout()

fig.savefig('fp_station_pos.png', dpi=300)
```
