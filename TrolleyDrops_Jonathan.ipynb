```python
def plt_unix_time_to_CST(ax):
    ax.locator_params(axis='x', nbins=5)
    xticks = ax.get_xticks()
    ax.set_xticklabels([pd.to_datetime(tm, unit='s').tz_localize('UTC').tz_convert('US/Central').strftime('%Y-%m-%d\n %H:%M:%S %Z')
                        for tm in xticks], rotation=0, fontdict={'size':10, 'family':'serif'})

def difference(y, avgt):      ##Compresses an array by averaging over the given averaging time to "bin" the data, difference is a misleading name
    t = np.arange(int(avgt) , len(y), int(avgt))
    N = len(t)
    Sum = 0
    avgdiff = np.array([])
    for i in range(len(y)):
        Sum += y[i]
        if i % avgt == 0 and i > 0:
            avg = Sum/avgt
            Sum = 0
            avgdiff = np.append(avgdiff, avg)
            
        
    return (t,avgdiff)

def MovingAvg(x, bin_size):
    avg = np.array([])
    for i in np.arange(len(x) - bin_size + 1):
    #while (i <= len(x) - bin_size):
        Sum = 0
        j = 0
        for j in np.arange(bin_size):
        #while (j < bin_size):
            Sum += x[i + j]
            #j += 1
        avg = np.append(avg, Sum)
        if(i % 86400 == 0):
            print(str(i) + "iterations done")
        i +=1
    return avg/bin_size

def unique_perms(series):
#https://stackoverflow.com/questions/4250125/generate-permutations-of-list-with-repeated-elements
    return {"".join(p) for p in permutations(series)}

def StringToNumpy(string):
    #This is only for string of ints with no separator
    # i.e. something like 1234 turns into [1,2,3,4]
    #used to take in unique_perms output
    arr = np.zeros(len(string), dtype = 'int')
    for i in range(len(string)):
        arr[i] = int(string[i])
    return arr

def DropsToString(nProbes, nDrops):
    #Makes a string of 0s and 1s for the unique_perm fxn
    dropStr = ''
    for i in range(nProbes):
        if i < nDrops:
            dropStr += '0'
        else:
            dropStr += '1'
    return dropStr

def DropCombinations(nProbes, nDrops):
    #Array of all possible drop arrays
    combStr = list(unique_perms(DropsToString(nProbes, nDrops)))
    numberOfCombinations = len(combStr)
    possibleDrops = np.zeros((numberOfCombinations, nProbes))
    for i in range(numberOfCombinations):
        possibleDrops[i] = StringToNumpy(combStr[i])
    return possibleDrops.astype('int')

def DropsToLabel(Drops):
    label = "Dropped probes: "
    for i in range(len(Drops)):
        if Drops[i] == 0:
            label += str(i) +", "
    return label[:-2]
```
