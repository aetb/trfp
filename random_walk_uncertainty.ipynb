```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def random_walk(N,A):
    output = np.zeros(N)
    for i in range(1,N):
        output[i] = output[i-1] + np.random.normal(scale=A)
    return output

def brownian_bridge(random_walk):
    linear = random_walk[-1]*np.linspace(0,1,len(random_walk))
    return random_walk-linear
```

```python
fig1, ax1 = plt.subplots(1,1)

x1 = random_walk(1000,1)
y1 = brownian_bridge(x1)
ax1.axhline(0, color='black')
ax1.plot(x1)

fig1.set_size_inches(8,4)
fig1.tight_layout()

fig2, ax2 = plt.subplots(1,1)

ax2.axhline(0, color='black')
ax2.plot(x1)
ax2.plot(y1)

fig2.set_size_inches(8,4)
fig2.tight_layout()

```

```python
trials = 20000
length = 100

means_rw = np.zeros(trials)
means_bb = np.zeros(trials)
lasts_rw = np.zeros(trials)

for i in range(trials):
    x = random_walk(length, 1)
    y = brownian_bridge(x)
    means_rw[i] = np.mean(x)
    means_bb[i] = np.mean(y)
    lasts_rw[i] = x[-1]

# print np.mean(means_rw)
print np.var(lasts_rw)
print np.var(means_rw)
print np.var(lasts_rw)/np.var(means_rw)

# print np.mean(means_bb)
print np.var(lasts_rw)
print np.var(means_bb)
print np.var(lasts_rw)/np.var(means_bb)
```
