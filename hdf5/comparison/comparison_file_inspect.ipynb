```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def plt_unix_time_to_CST(ax):
    ax.locator_params(axis='x', nbins=10)
    xticks = ax.get_xticks()
    ax.set_xticklabels([pd.to_datetime(tm, unit='s').tz_localize('UTC').tz_convert('US/Central').strftime('%Y-%m-%d\n %H:%M:%S %Z')
                          for tm in xticks], rotation=30, fontdict={'size':12, 'family':'serif'})

def plt_set_labels(ax, x_label, y_label, title):
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.set_title(title)
```

```python
filename = '60hr_vtm_1_comparison'

data = np.loadtxt(filename+'.csv', delimiter=',', skiprows=1)

fig, axs = plt.subplots(3,1)
axs[0].plot(data[:,0], data[:,1], '.')
plt_set_labels(axs[0], '', 'm1 (Hz)', filename)
axs[1].plot(data[:,0], data[:,2], '.')
plt_set_labels(axs[1], '', 'm2 (Hz)', '')
axs[2].plot(data[:,0], data[:,3], '.')
plt_set_labels(axs[2], '', 'm3 (Hz)', '')

axs[0].set_ylim(np.median(data[:,1]) - 3 * np.std(data[:,1]), np.median(data[:,1]) + 3 * np.std(data[:,1]))
axs[1].set_ylim(np.median(data[:,2]) - 3 * np.std(data[:,2]), np.median(data[:,2]) + 3 * np.std(data[:,2]))
axs[2].set_ylim(np.median(data[:,3]) - 3 * np.std(data[:,3]), np.median(data[:,3]) + 3 * np.std(data[:,3]))

for ax in axs:
    plt_unix_time_to_CST(ax)

fig.set_size_inches(12,9)
fig.tight_layout()
```
