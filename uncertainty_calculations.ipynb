```python
# trolley uncertainty table

uncertainties = {'freq_extract':0.3,
                'trolley_probe_calib':0.74,
                'azi_pos':0.3,
                'trolley_spikes':0.3,
                'other':1.2}

total = 0
for source in uncertainties:
    total += uncertainties[source]**2


for xy_position in [.6, 2.5, 0.2]:
    print np.sqrt(total + xy_position**2)
    print np.sqrt(total + xy_position**2)/61.79*1000
    print '\n'
```
