```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import numpy as np
import pandas as pd

import gm2
import trfp
import helper_function_candidates as helper
import plotting_functions as plt2

import allantools
```

```python
moment_df = helper.import_fp_moment_runs("9day_fixed_probe_runs_1.h5", [4141])

interp_df = pd.read_hdf("9day_fixed_probe_runs_1.h5", key="run_4141_interp_df")
```

```python
# make Allan plots for fp0 and for st0,m1 (good probe, good station)

fig, axs = plt.subplots(2,1)
plt.sca(axs[0])

taus, adev, _, _ = allantools.mdev(interp_df['fp1'].values, data_type='freq', rate=1, taus='all')
plt.loglog(taus, adev, '.', color='navy')

plt.sca(axs[1])
taus, adev, _, _ = allantools.mdev(moment_df['st8,m1'].values, data_type='freq', rate=1, taus='all')
plt.loglog(taus, adev, '.', color='navy')

fig.set_size_inches(12,8)
fig.tight_layout()
```
