```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('darkgrid')

fig, ax = plt.subplots(1,1)

x = np.linspace(0,1,100)
y1 = x[0:50]
y2 = x[50:]
y3 = 1-x[50:]

ax.plot(x[0:50], y1, color='C0', label = 'forward')
ax.plot(x[50:], y2, '--', color='C0')
ax.plot(x[50:], y3, color='C2', label = 'backward')

y13 = np.append(y1, y3)

ax.fill_between(x, y13, color='C1', alpha=.5)

ax.plot(1,1,'.', color='red', markersize=12, label = 'sync offset')

ax.legend()

ax.set_xticks([0,0.25,0.5,0.75,1])
ax.set_xticklabels(['0', '', 'T/2', '', 'T'])

ax.set_yticks([0,0.25,0.5,0.75,1])
ax.set_yticklabels(['0', '', '$\Delta M^2/2$', '', '$\Delta M^2$'])

ax.set_xlabel('Interpolation time from trolley run (arb.)')
ax.set_ylabel('Uncertainty $\sigma^2$ (arb.)')

fig.set_size_inches(6,4)
fig.tight_layout()
```
